```python
import pandas as pd
df = pd.read_csv('mark.csv')
print(df.head())
```

```python
df['Total']=df[['Math','Science','English']].sum(axis=1)
print(df)
```

```python
df['Total']=df[['Math','Science','English']].sum(axis=1)
df[['name','Math','Science','English','Total']].head()
```

```python
# Function to assign grades
```

```python
def assign_grade(Total):
    if Total >= 250:
        return 'A'
    elif Total >= 200:
        return 'B'
    elif Total >= 150:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['Total'].apply(assign_grade)

print(df[['name','Total','Grade']]) 

# df[['name','Total','Grade']].head()  #prints first 5 students

# df[['name','Total','Grade']].tail()  #prints last 5 students

```

```python
# list of grade A students
grade_a_students = df[df['Grade']== 'A']
print(grade_a_students)
```

```python
# list of grade B students
grade_b_students = df[df['Grade']== 'B']
print(grade_b_students)
```

```python
# list of grade C students
grade_c_students = df[df['Grade']== 'C']
print(grade_c_students)
```

```python
# list of grade D students
grade_d_students = df[df['Grade']== 'D']
print(grade_d_students)
```

```python
# length of Grade D Students
grade_d_students = df[df['Grade']== 'D']
print(len(grade_d_students))
```

```python
df.to_csv("result.csv",index=False)
print("File is Created")
```
