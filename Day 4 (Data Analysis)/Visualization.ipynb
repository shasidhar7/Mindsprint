```python
import matplotlib.pyplot as plt
print(plt.__package__)  ##See the package name
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
plt.plot(x,y, marker='o', linestyle = '-', color= 'b')

plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Lable")
plt.grid(True)
plt.show()
 
```

```python
plt.figure(figsize=(10,6))
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.scatter(x,y,marker='*',color='r')
plt.xlabel("X-axis")
plt.ylabel("Y-lable")
plt.title("Line Plot")
plt.grid()
plt.show()
 
```

```python

categories = ['A','B', 'C', 'D']
values = [5, 7, 3, 8]
 
plt.bar(categories, values, color='green')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar graph")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random numbers
data = np.random.randn(1000)
plt.hist(data, bins=10, color='purple', edgecolor='black')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```
